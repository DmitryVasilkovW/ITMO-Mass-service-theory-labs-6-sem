## Моделирование сети массового обслуживания

**Тема:** банк

### Описание модели

#### Параметры сети
- **Входящий поток заявок (λ)**: 10 заявок/час
- **Интенсивность обслуживания (μ)**: 3 заявки/час
- **Максимальное число агентов**: 5
- **Минимальное число агентов**: 1
- **Время моделирования**: 1000 часов

#### Алгоритм работы агентов
1. Агенты динамически подключаются и отключаются во времени.
2. Если агент подключается, увеличивается количество доступных серверов.
3. Если агент отключается, уменьшается количество серверов.
4. Если все серверы заняты, заявки попадают в очередь.
5. Если заявка ждёт слишком долго, она теряется.
