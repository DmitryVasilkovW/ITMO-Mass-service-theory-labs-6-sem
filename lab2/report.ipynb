## 2. Описание модели
Система представляет собой многоканальную СМО с ожиданием (M/M/n/∞), где:
- **λ** — интенсивность входящего потока заявок (заявок/час).
- **μ** — интенсивность обслуживания заявок одним каналом (заявок/час).
- **n** — количество каналов обслуживания.
- Длина очереди не ограничена (∞).

**Основные допущения:**
- Входящий поток заявок — пуассоновский.
- Время обслуживания — экспоненциальное.
- Заявки обслуживаются в порядке очереди (FIFO).

**Формулы для расчета характеристик:**
1. **Вероятность простоя системы (p₀):**
   $$
   p_0 = \left[ \sum_{k=0}^{n-1} \frac{(\lambda / \mu)^k}{k!} + \frac{(\lambda / \mu)^n}{n! \cdot (1 - \rho)} \right]^{-1}, \quad \rho = \frac{\lambda}{n \mu}
   $$
2. **Вероятность того, что заявка будет ждать в очереди (p_que):**
   $$
   p_{\text{que}} = \frac{(\lambda / \mu)^n}{n! \cdot (1 - \rho)} \cdot p_0
   $$
3. **Среднее число заявок в очереди (Lq):**
   $$
   L_q = \frac{p_{\text{que}} \cdot \rho}{1 - \rho}
   $$
4. **Среднее время ожидания в очереди (Wq):**
   $$
   W_q = \frac{L_q}{\lambda}
   $$
5. **Среднее время пребывания заявки в системе (W):**
   $$
   W = W_q + \frac{1}{\mu}
   $$
6. **Среднее число заявок в системе (L):**
   $$
   L = \lambda \cdot W
   $$
7. **Коэффициент загрузки системы (ρ):**
   $$
   \rho = \frac{\lambda}{n \mu}
   $$

## 4. Анализ результатов

### Влияние количества каналов (n)
При увеличении количества каналов:
- Среднее время ожидания (Wq) и длина очереди (Lq) уменьшаются.
- Например, при n=4: Wq=0.3288 часа ≈20 мин), при n=6: Wq≈0.02 часа (≈1.2 мин).
- Графики показывают резкое снижение Wq и Lq при добавлении каналов.

### Влияние интенсивности обслуживания (μ)
При увеличении интенсивности обслуживания:
- Среднее время ожидания (Wq) и длина очереди (Lq) уменьшаются.
- Например, при μ=3: Wq=0.3288 часа, при μ=5: Wq≈0.02 часа.
