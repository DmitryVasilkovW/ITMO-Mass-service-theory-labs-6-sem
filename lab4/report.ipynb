## Анализ системы массового обслуживания с приоритетами

**Тема:** банковское обслуживание "одно окошко"

##  Описание модели

В данной работе исследуется **система массового обслуживания (СМО)** с **абсолютным приоритетом**, где:
- **Высокоприоритетные заявки** $$ \lambda_1 $$ обслуживаются в первую очередь.
- **Низкоприоритетные заявки** $$ \lambda_2 $$ обслуживаются только при отсутствии заявок с более высоким приоритетом.

###  **Исходные параметры системы**
- Интенсивность потока **высокоприоритетных заявок**: $$ \lambda_1 = 3 $$ заявки/час
- Интенсивность потока **низкоприоритетных заявок**: $$ \lambda_2 = 5 $$ заявок/час
- Интенсивность обслуживания: $$ \mu = 10 $$ заявок/час

##  Формулы для расчёта характеристик системы

1. **Коэффициент загрузки системы**
   $$
   \rho = \frac{\lambda_1 + \lambda_2}{\mu}
   $$

2. **Среднее время ожидания в очереди**
   - Для высокоприоритетных заявок:
     $$
     W_q^{(1)} = \frac{\lambda_2}{\mu (\mu - \lambda_1 - \lambda_2)}
     $$
   - Для низкоприоритетных заявок:
     $$
     W_q^{(2)} = \frac{\lambda_1 + \lambda_2}{\mu (\mu - \lambda_1 - \lambda_2)}
     $$

3. **Среднее время пребывания в системе**
   - Для высокоприоритетных заявок:
     $$
     W^{(1)} = \frac{1}{\mu}
     $$
   - Для низкоприоритетных заявок:
     $$
     W^{(2)} = W_q^{(2)} + \frac{1}{\mu}
     $$

4. **Вероятность ожидания в очереди** для низкоприоритетных заявок:
   $$
   P_{\text{wait}}^{(2)} = \frac{\rho_2}{1 - \rho_1}
   $$

где:
- $$\rho_1 = \frac{\lambda_1}{\mu}$$ – загрузка системы высокоприоритетными заявками
- $$ \rho_2 = \frac{\lambda_2}{\mu} $$ – загрузка системы низкоприоритетными заявками

## Анализ результатов

1. **Высокоприоритетные заявки обрабатываются немедленно**, поэтому их среднее время ожидания в очереди равно 0.
2. **Низкоприоритетные заявки вынуждены ждать**, так как высокоприоритетные заявки блокируют часть ресурсов.
3. **Чем больше интенсивность потока низкоприоритетных заявок**, тем дольше их время ожидания.
4. **При слишком высокой нагрузке $$ \lambda_1 + \lambda_2 \approx \mu $$ система становится перегруженной**, что приводит к резкому увеличению времени ожидания.
