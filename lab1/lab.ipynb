## Выводы

Из проведённых экспериментов можно сделать следующие выводы:

- **Общее число заявок:**
  При увеличении λ общее число заявок растёт. Однако, из-за ограниченности системы (одно окно), не все заявки могут быть обслужены.

- **Вероятность отказа:**
  С увеличением интенсивности входящего потока (λ) растёт и экспериментальная вероятность отказа. Это связано с тем, что окно обслуживания не успевает принимать все поступающие заявки, и система начинает отказывать в обслуживании.

- **Сравнение с теоретическим значением:**
  Теоретическая вероятность отказа рассчитывается по формуле:
$$
P_{\text{отказа}} = \frac{\rho}{1+\rho}, \quad \text{где } \rho = \frac{\lambda}{\mu}.
$$
. Результаты эксперимента хорошо коррелируют с теоретическими значениями.

- **Коэффициент загрузки:**
  При низком значении λ окно простаивает большую часть времени, а при высоком – почти постоянно занято. Это влияет на эффективность работы системы и качество обслуживания.

Таким образом, моделирование позволяет увидеть, как изменения параметров системы (λ и μ) влияют на её характеристики и производительность.
